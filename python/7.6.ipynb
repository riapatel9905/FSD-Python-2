```python
import requests
response=requests.get("https://api.github.com/rate_limit")
data=response.json()
print(data)
print("Limit",data["rate"]["limit"])
print("Used",data["rate"]["used"])
print("Remaining",data["rate"]["remaining"])
print("Reset",data["rate"]["reset"])
```

```python
import requests
for i in range(70):
    r=requests.get("https://api.github.com/users/octocat")
    print(i+1,r.status_code)
    if r.status_code==403:
        data=r.json()
        if "API rate limit exceeded" in data.get("message",""):
            print("rate limit exceeded")
            print(data["message"])
            break
```

```python
from datetime import datetime
import requests
r=requests.get("https://api.github.com/rate_limit")
data=r.json()
reset_stamp=data['rate']['reset']
reset_time=datetime.fromtimestamp(reset_stamp).strftime("%y-%m-%d-%H-%M-%S")
print(reset_time)
```
