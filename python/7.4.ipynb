```python
import requests,json
def get_coordinates():
    api_key="58b6386094d4435d3b68cf35d4ecf8b8"
    city=input("Enter City Name: ")
    request_url=f"http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=5&appid={api_key}"
    response=requests.get(request_url)
    coordinates=response.json()
    print(json.dumps(coordinates,indent=5))
    lattitude=coordinates[0]['lat']
    longitude=coordinates[0]['lon']
    return lattitude,longitude,city
lat,lon,city=get_coordinates()
print(city, ":Lattitude & Longitude")
print("------------------------------")
print("Lattitude: ",lat)
print("Longitude: ",lon)
```

```python
import requests,json
def current_weather_data():
    api_key="58b6386094d4435d3b68cf35d4ecf8b8"
    request_url=f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url)
    weather_data=response.json()
    print(json.dumps(weather_data,indent=5))
    description=weather_data['weather'][0]['description']
    temperature=weather_data['main']['temp']
    pressure=weather_data['main']['humidity']
    wind_speed=weather_data['wind']['speed']
    visibility=weather_data['visibility']
    print("Current Weather Data")
    print("----------------------")
    print("Weather Description",description)
    print("Temperature",temperature)
    print("Pressure",pressure)
    print("Wind Speed",wind_speed)
    print("Visibility",visibility)
current_weather_data()
```

```python
def air_pollution():
    api_key="58b6386094d4435d3b68cf35d4ecf8b8"
    url=f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(url)
    a=response.json()
    print(json.dumps(a,indent=5))
    quality_index={1:"Good",2:"Fair",3:"Moderate",4:"Poor",5:"Very Poor"}
    index=a['list'][0]['main']['aqi']
    print("AQI",quality_index[index])
air_pollution()
```

```python
import requests,json
def weather_forecast():
    api_key="58b6386094d4435d3b68cf35d4ecf8b8"
    city=input("Enter City Name: ")
    request_url=f"http://api.openweathermap.org/geo/1.0/direct?q={city}&limit=5&appid={api_key}"
    response=requests.get(request_url)
    coordinates=response.json()
    #print(json.dumps(coordinates,indent=5))
    lattitude=coordinates[0]['lat']
    longitude=coordinates[0]['lon']
    request_url1=f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url1)
    five_day=response.json()
    print(json.dumps(five_day,indent=5))
    return five_day
five_day=weather_forecast()
```

```python
import pandas as pd
data={"Date_Time":[],"Temperature":[],"Pressure":[],"Humidity":[],"Wind Speed":[],"Visibility":[],"Description":[]}
for i in five_day['list']:
    data["Date_Time"].append(i['dt_txt'])
    data["Temperature"].append(i['main']['temp'])
    data["Pressure"].append(i['main']["pressure"])
    data["Humidity"].append(i['main']["humidity"])
    data["Description"].append(i['weather'][0]["description"])
    data["Wind Speed"].append(i['wind']["speed"])
    data["Visibility"].append(i["visibility"])
df=pd.DataFrame(data)
df
```
