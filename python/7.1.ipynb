```python
from bs4 import BeautifulSoup
import requests
url="https://subslikescript.com/movies"
result=requests.get(url)
content=result.text
```

```python
soup=BeautifulSoup(content,"html.parser")
print(soup.prettify())
```

```python
len(soup.find_all('a',href=True))
```

```python
box=soup.find('article',class_="main-article")
print(box)
```

```python
box.find_all('a')
```

```python
links=[]
for links in box.find_all('a',href=True):
    print(links['href'])
```

```python
links=[]
for link in box.find_all('a',href=True):
    links.append(link['href'])
print(links)
    
```

```python
for i in links:
    print("https://subslikescript.com"+i)
```

```python
urls=[]
for i in links:
    urls.append("https://subslikescript.com"+i)
```

```python
print(urls)
```

```python
for i in urls:
    result=requests.get(i)
    content=result.text
    soup=BeautifulSoup(content,"html.parser")
    box=soup.find("article",class_="main-article")
    title=box.find('h1').get_text()
    script=box.find('div',class_="full-script").get_text()
    try:
        with open(f"{title}.txt","w",encoding='utf-8') as file:
            file.write(script)
    except Exception as e:
        print(e)
                      
```
