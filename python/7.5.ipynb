```python
import requests
import json
url="https://jsonplaceholder.typicode.com/posts/1"
response=requests.get(url)
data=response.json()
with open('post.json','w') as f:
    json.dump(data,f,indent=5)
print("JSON file saved")
```

```python
with open('post.json','r') as f:
    load_data=json.load(f)
print(load_data)
```

```python
import requests
import json
url="https://jsonplaceholder.typicode.com/users"
response=requests.get(url)
data=response.json()
for i in data:
    name=i['name']
    uname=i['username']
    phone=i['phone']
    email=i['email']
    city=i['address']['city']
    lat=i['address']['geo']['lat']
    lng=i['address']['geo']['lng']

    print(name)
    print(uname)
    print(phone)
    print(email)
    print(city)
    print(lat)
    print(lng)
    print("---------------------------")
```

```python
import requests
import json
url="https://jsonplaceholder.typicode.com/users"
response=requests.get(url)
data=response.json()
rows=[]
for i in data:
    name=i['name']
    uname=i['username']
    phone=i['phone']
    email=i['email']
    city=i['address']['city']
    lat=i['address']['geo']['lat']
    lng=i['address']['geo']['lng']
    d={'Name':name,"UserName":uname,"Phone":phone,"email":email,"city":city,"lat":lat,"lon":lng}
    rows.append(d)
df=pd.DataFrame(rows)
df
```

```python
import requests
import json
url="https://jsonplaceholder.typicode.com/posts"
page=1
while page<=3:
    params={'_page':page,'_limit':10}
    response=requests.get(url,params=params)
    data=response.json()
    for i in data:
        print(i['id'],"-",i['title'])
    page+=1
```
