```python
from bs4 import BeautifulSoup
import requests
url="https://www.politifact.com/factchecks/"
result=requests.get(url)
#result.status_code
content=result.text
soup=BeautifulSoup(content,'html.parser')
#print(soup.prettify())
```

```python
data = soup.find_all('li',class_='o-listicle__item')
#data
```

```python
#data[0]
```

```python
source=[]
for i in data:
    source.append(i.find('div',class_='m-statement__meta').find('a').text.strip())
print(source)
```

```python
statement=[]
for i in data:
    statement.append(i.find('div',class_="m-statement__quote").text.strip())
print(statement)
```

```python
date=[]
for i in data:
    date.append(i.find('footer',class_="m-statement__footer").text.split(" • ")[-1].strip())
print(date)
```

```python
links=[]
root="https://www.politifact.com"
for i in data:
    d=i.find("div",class_="m-statement__quote").find('a')['href'].strip()
    links.append(root+d)
print(links)
```

```python
import pandas as pd
df=pd.DataFrame({'statement':statement,'source':source,'date':date,'link':links})
df
```

```python
df.to_excel('politifact.xlsx')
```

```python
import requests
from bs4 import BeautifulSoup
upperframe=[]
for page in range(1,5):
    url="https://www.politifact.com/factchecks/?page="+str(page)
    result=requests.get(url)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    frame=[]
    data=soup.find_all('li',class_='o-listicle__item')
    for i in data:
        source=i.find('div',class_="m-statement__meta").find('a').text.strip()
        statement=i.find('div',class_="m-statement__quote").text.strip()
        date=i.find('footer',class_="m-statement__footer").text.split(' • ')[-1].strip()
        links="https://www.politifact.com"+i.find('div',class_="m-statement__quote").find('a')['href'].strip()
        frame.append((statement,source,date,links))
    upperframe.extend(frame)
print(len(upperframe))
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(open('top250.html','r',encoding="utf-8"))
#print(soup.prettify())
```

```python
movie=soup.find_all('td',class_="titleColumn")
movie[0]
```

```python
len(movie)
```

```python
title=[]
for i in movie:
    title.append(i.find('a').text.strip())
print(title)
```

```python
year=[]
for i in movie:
    year.append(i.find('span',class_='secondaryInfo').text[1:-1].strip())
print(year)
```

```python
rating=[]
for i in soup.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(i.text.strip())
print(rating)                       
```

```python
import pandas as pd
df=pd.DataFrame({'Movie Title':title,'Release Year':year,'IMDB Rating':rating})
df
```
